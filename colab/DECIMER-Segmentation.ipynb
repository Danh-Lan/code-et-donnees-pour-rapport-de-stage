```python
import numpy as np
import time
import cv2
import os
import matplotlib.pyplot as plt
%matplotlib inline

from decimer_segmentation import segment_chemical_structures, segment_chemical_structures_from_file
```

```python
# display function to show image on Jupyter
def display_img(img,cmap=None):
    fig = plt.figure(figsize = (8,8))
    plt.axis(False)
    AX = fig.add_subplot(111)
    AX.imshow(img,cmap)
```

Run DECIMER Segmentation with the mask expansion

```python
img = cv2.imread("drive/MyDrive/YOLOv3_files/test/test (9).jpg")
#display_img(img)

segments = segment_chemical_structures(np.array(img),
                                       expand=False,
                                       visualization=True)
```

Run DECIMER Segmentation directly on a file

```python
path = os.path.abspath("drive/MyDrive/YOLOv3_files/test/test (5).jpg")
segments = segment_chemical_structures_from_file(path,
                                                 expand=False,
                                                 poppler_path=None)
#Show the image from each segment
for i in range (len(segments)):
    display_img(segments[i])
```
